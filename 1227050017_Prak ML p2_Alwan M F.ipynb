## IMPORT LIBRARY

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## IMPORT DATASET

```python
dataset = pd.read_csv(r"D:\Kuliah Semester 6\Praktikum Pembelajaran Mesin\2\tips.csv")
```

## Ambil Nilai dari Kolom 

```python
X = dataset.iloc[:, 2].values
Y = dataset.iloc[:, 1].values
kategori = dataset.iloc[:, 2].values
```

```python

warna = ['red' if k == "Male" else 'blue' for k in kategori]
```

```python
x_min, x_max = X.min(), X.max()
y_min, y_max = Y.min(), Y.max()
```

# 1. Scatter

```python

plt.scatter(X, Y, color=warna)
plt.xlim(X.min(), X.max())
plt.ylim(Y.min(), Y.max())
plt.xlabel('Gender')
plt.ylabel('Tips')
plt.title('Sebaran Data')
plt.show()
```

## 2. Bar Chart

```python
plt.bar(X, Y, color=warna)
plt.xlim(X.min(), X.max())
plt.ylim(Y.min(), Y.max())
plt.xlabel('Gender')
plt.ylabel('Tips')
plt.title('Sebaran Data')
plt.show()
```

## 3. BOX PLOT

```python
sns.boxplot(dataset)
plt.title('Box Plot')
plt.show()
```

## 4. Pie Chart

```python
grouped = dataset.groupby('sex')['tip'].sum()

# Menghitung persentase
total_tip = grouped.sum()
percentages = (grouped / total_tip) * 100

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(grouped, labels=[f"{sex} ({percentage:.1f}%)" for sex, percentage in zip(grouped.index, percentages)],
        autopct='%1.1f%%', colors=warna, startangle=140)
plt.title('Distribusi Tip Berdasarkan Gender')
plt.show()
```
